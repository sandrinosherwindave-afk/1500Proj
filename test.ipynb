```python
import tkinter as tk

def check_answer():
    # Retrieve the text typed into the entry bar
    user_answer = answer_entry.get()
    
    # Check if the answer is correct (ignoring extra spaces)
    if user_answer.strip() == "12":
        show_minus_one()
        # Clear the entry box so it's empty again
        answer_entry.delete(0, tk.END)

def show_minus_one():
    # Update the label to show "-1"
    feedback_label.config(text="-1")
    
    # Schedule the text to hide after 1000 milliseconds (1 second)
    root.after(1000, hide_text)

def hide_text():
    # Clear the text from the label
    feedback_label.config(text="")

# Set up the main window
root = tk.Tk()
root.title("Q&A App")
root.geometry("300x250")

# Create the question label
question_label = tk.Label(root, text="What is 5 + 7?", font=("Helvetica", 14))
question_label.pack(pady=20)

# Create the text entry bar for the answer
answer_entry = tk.Entry(root, font=("Helvetica", 14), justify="center")
answer_entry.pack(pady=10)

# Create the submit button and link it to the check_answer function
submit_button = tk.Button(root, text="Submit", command=check_answer)
submit_button.pack(pady=10)

# Create a label to hold the "-1" text, initially empty
feedback_label = tk.Label(root, text="", font=("Helvetica", 16), fg="red")
feedback_label.pack()

# Start the application
root.mainloop()
```

```python
import tkinter as tk
from drawings import *

class ArrowMenu:
    def __init__(self, root):
        self.root = root
        self.menu_options = {
            "Start Game": self.start_game,
            "Settings": None,
            "Quit": self.root.destroy
        }
        
        self.items = list(self.menu_options.keys())
        self.current_idx = 0
        
        self.lb = tk.Listbox(
            root, 
            font=("Courier", 16, "bold"),
            bg="black",          # Background color
            fg="white",          # Text color
            bd=0,                # Border width
            highlightthickness=0, # Remove focus border
            selectborderwidth=0,  # Remove selection border
            activestyle="none",   # Remove dotted underline
            selectbackground="black", # Hide blue highlight
            width=20, 
            height=10
        )
        self.lb.pack(pady=20)
        
        self.update_list()
        
        # Bind keys
        self.root.bind("<Up>", self.move_up)
        self.root.bind("<Down>", self.move_down)
        self.root.bind("<Return>", self.execute_selection)
        
    def start_game(self):
        self.lb.pack_forget()
        nick_player = tk.Label(root,
                                   text = nick_model,
                                   font = ("Courier New", 14),
                                   fg = "white",
                                   bg = "black",
                                   justify= tk.LEFT,
                                   )
        nick_player.place(x= 660, y = 192)
        
    

    def update_list(self):
        self.lb.delete(0, tk.END)
        for i, item in enumerate(self.items):
            if i == self.current_idx:
                self.lb.insert(tk.END, f"→ {item}")
                self.lb.itemconfig(i, fg="blue")
            else:
                self.lb.insert(tk.END, f"  {item}")
        
        self.lb.selection_clear(0, tk.END)
        self.lb.selection_set(self.current_idx)

    def move_up(self, event):
        if self.current_idx > 0:
            self.current_idx -= 1
            self.update_list()

    def move_down(self, event):
        if self.current_idx < len(self.items) - 1:
            self.current_idx += 1
            self.update_list()

    def execute_selection(self, _):
        # 2. Get the key name, then find and run the function
        key = self.items[self.current_idx]
        action = self.menu_options[key]
        action() # This calls the function

root = tk.Tk()
root.title("Quiz Game")
root.state('zoomed')
root.configure(background="black")
app = ArrowMenu(root)
root.mainloop()
```

```python
import tkinter as tk
import random

# -----------------------
# GAME CONSTANTS
# -----------------------
PLAYER_MAX_HP = 50
ENEMY_MAX_HP = 50

# Built-in questions (no AI)
QUESTIONS = {
    "easy": [
        {"q": "What is 2 + 2?", "a": "4"},
        {"q": "What color is the sky?", "a": "blue"},
        {"q": "Capital of France?", "a": "paris"}
    ],
    "medium": [
        {"q": "12 x 12?", "a": "144"},
        {"q": "Who wrote Romeo and Juliet?", "a": "shakespeare"}
    ],
    "hard": [
        {"q": "Square root of 256?", "a": "16"},
        {"q": "Red Planet?", "a": "mars"}
    ]
}

DIFFS = ["easy", "medium", "hard"]

DIFF_CONFIG = {
    "easy": {"dmg": (15, 25), "penalty": 8},
    "medium": {"dmg": (30, 45), "penalty": 14},
    "hard": {"dmg": (50, 70), "penalty": 22}
}

# -----------------------
# GAME STATE
# -----------------------
state = {
    "player_hp": PLAYER_MAX_HP,
    "enemy_hp": ENEMY_MAX_HP,
    "current_diff": "easy",
    "current_answer": "",
    "skip_count": 0
}

# -----------------------
# FUNCTIONS
# -----------------------
def update_ui():
    player_label.config(text=f"Player HP: {state['player_hp']}/{PLAYER_MAX_HP}")
    enemy_label.config(text=f"Enemy HP: {state['enemy_hp']}/{ENEMY_MAX_HP}")


def set_question(diff):
    state["current_diff"] = diff
    q = random.choice(QUESTIONS[diff])
    state["current_answer"] = q["a"]

    question_label.config(text=f"{diff.upper()} QUESTION:\n{q['q']}")
    entry.delete(0, tk.END)


def submit_answer():
    user = entry.get().strip().lower()
    if not user:
        return

    cfg = DIFF_CONFIG[state["current_diff"]]

    if user == state["current_answer"]:
        dmg = random.randint(*cfg["dmg"])
        state["enemy_hp"] = max(0, state["enemy_hp"] - dmg)
        result_label.config(text=f"Correct! You dealt {dmg} damage!")
    else:
        state["player_hp"] = max(0, state["player_hp"] - cfg["penalty"])
        result_label.config(text=f"Wrong! Answer: {state['current_answer']}\nYou took {cfg['penalty']} damage!")

    update_ui()
    check_game()


def skip_question():
    state["skip_count"] += 1

    if state["skip_count"] >= 3:
        dmg = 15
        state["player_hp"] = max(0, state["player_hp"] - dmg)
        result_label.config(text=f"You used all 3 skips!\nMonster hits you for {dmg} damage!")
        update_ui()
        check_game()
        return

    # Cycle difficulty: easy -> medium -> hard -> easy
    current_index = DIFFS.index(state["current_diff"])
    next_diff = DIFFS[(current_index + 1) % len(DIFFS)]

    result_label.config(text=f"Skip used ({state['skip_count']}/3)\nNew difficulty: {next_diff.upper()}")
    set_question(next_diff)


def check_game():
    if state["enemy_hp"] <= 0:
        result_label.config(text="YOU WIN!")
        disable_buttons()
    elif state["player_hp"] <= 0:
        result_label.config(text="YOU LOST!")
        disable_buttons()


def disable_buttons():
    submit_btn.config(state="disabled")
    skip_btn.config(state="disabled")


def enable_buttons():
    submit_btn.config(state="normal")
    skip_btn.config(state="normal")


def restart_game():
    state["player_hp"] = PLAYER_MAX_HP
    state["enemy_hp"] = ENEMY_MAX_HP
    state["skip_count"] = 0
    state["current_diff"] = "easy"

    result_label.config(text="")
    update_ui()
    enable_buttons()
    set_question("easy")

# -----------------------
# UI SETUP
# -----------------------
root = tk.Tk()
root.title("Battle Quiz Game")
root.geometry("400x400")

player_label = tk.Label(root, font=("Arial", 12))
player_label.pack()

enemy_label = tk.Label(root, font=("Arial", 12))
enemy_label.pack()

question_label = tk.Label(root, wraplength=300, font=("Arial", 11))
question_label.pack(pady=10)

entry = tk.Entry(root)
entry.pack()

submit_btn = tk.Button(root, text="Submit", command=submit_answer)
submit_btn.pack(pady=5)

skip_btn = tk.Button(root, text="Skip (3 max)", command=skip_question)
skip_btn.pack(pady=5)

result_label = tk.Label(root, fg="red", font=("Arial", 10))
result_label.pack(pady=10)

restart_btn = tk.Button(root, text="Restart", command=restart_game)
restart_btn.pack(pady=10)

# Start game
update_ui()
set_question("easy")

root.mainloop()
```
